```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

Univariate Normal
$$
p(x;\mu,\sigma^2)=\frac{1}{\sqrt{2\pi\sigma^2}} \exp\biggl( \frac{-1}{2\sigma^2}(x-\mu)^2 \biggr)
$$

```python
class normal(object):
    def __init__(self, mu=0.0, sigma=1.0):
        self.mu = mu
        self.sigma = sigma
    def pdf(self, x):
        alpha = np.sqrt(2*3.14)*self.sigma
        beta = -0.5*self.sigma**(-0.5)
        gamma = (x-self.mu)**2
        return alpha**(-1)*np.exp(beta*gamma)
    def logpdf(self, x):
        alpha = np.log(np.sqrt(2*3.14)*self.sigma)
        beta = -0.5*self.sigma**(-0.5)
        gamma = (x-self.mu)**2
        return alpha+beta*gamma
```

```python
x = np.linspace(0,10,100).reshape(-1,1)
y = normal(5.0,1.0).pdf(x)
y_log = normal(5.0,1.0).logpdf(x)

plt.figure(figsize=(20,3))
plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(x,y_log)
plt.show()
```

```python
x = np.linspace(0,10,100).reshape(-1,1)
y = st.norm(5,1).pdf(x)
y_log = st.norm(5,1).logpdf(x)

plt.figure(figsize=(20,3))
plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(x,y_log)
plt.show()
```

Multivariate Normal
$$
p(x;\mu,\Sigma)= \frac{1}{(2\pi)^{\frac{n}{2}} \left\lvert \Sigma \right\rvert ^\frac{1}{2}} \exp\biggl( \frac{-1}{2}(x-\mu)^T \Sigma^{-1} (x-\mu) \biggr)
$$
