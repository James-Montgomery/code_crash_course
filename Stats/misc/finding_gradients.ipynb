```python
import numpy as np
import autograd
from autograd import grad
from autograd import numpy as auto_np
import scipy as sp
import matplotlib.pyplot as plt
import sympy as sym
```

```python
def f(x):
    return auto_np.sin(x)

gradient = grad(f)
double_gradient = grad(gradient)

x = np.linspace(-10,10,100)
y = f(x)
dy = [gradient(i) for i in x]
ddy = [double_gradient(i) for i in x]
```

```python
plt.plot(x,y)
plt.plot(x,dy)
plt.plot(x,ddy)
plt.show()
```

```python
interval = x[-1] - x[-2]
dy = np.gradient(y,interval)
ddy = np.gradient(dy,interval)
```

```python
plt.plot(x,y)
plt.plot(x,dy)
plt.plot(x,ddy)
plt.show()
```

```python
from scipy import optimize
```

```python
def f(x):
    return np.sin(x)

x = np.linspace(-10,10,100)
eps = np.sqrt(np.finfo(float).eps)
y = f(x)
g = lambda a: optimize.approx_fprime([a], f, [eps])
dy = [g(i) for i in x]
```

```python
plt.plot(x,y)
plt.plot(x,dy)
plt.show()
```
